```python
import ifcopenshell

def extract_walls_from_ifc(ifc_file_path):
    # Open the IFC file
    ifc_file = ifcopenshell.open(ifc_file_path)
    
    # Filter all IfcWall elements
    walls = ifc_file.by_type("IfcWall")
    
    wall_data = []

    for wall in walls:
        wall_guid = wall.GlobalId
        body_representation = None
        
        # Iterate through the representations of the wall
        for representation in wall.Representation.Representations:
            if representation.RepresentationType == "SweptSolid":
                body_representation = "SweptSolid"
                break
            elif representation.RepresentationType == "Brep":
                body_representation = "Brep"
                break
            elif representation.RepresentationType == "SurfaceModel":
                body_representation = "SurfaceModel"
                break
            elif representation.RepresentationType == "Tessellation":
                body_representation = "Tessellation"
                break
            # Add more RepresentationTypes here if needed

        wall_data.append([wall_guid, body_representation])

    return wall_data

# Path to the IFC file
ifc_file_path = "Hus28_test.ifc"

# Extract walls and their body representation types
walls = extract_walls_from_ifc(ifc_file_path)

# Print the results
for wall in walls:
    print(f"GUID: {wall[0]}, Body Representation: {wall[1]}")
```
