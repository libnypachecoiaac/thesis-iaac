```python
import ifcopenshell

# IFC-Datei öffnen
ifc_file = ifcopenshell.open("../Hus28_test.ifc")

# Die gesuchte IfcGuid / GlobalId
target_guid = "1wOE2MwBf03f9R8GpskFuS"

# Alle IfcWindows durchsuchen
for ifc_window in ifc_file.by_type("IfcWindow"):
    # Prüfen, ob die GlobalId des Fensters mit der gesuchten übereinstimmt
    if ifc_window.GlobalId == target_guid:
        print(f"Fenster mit GlobalId {target_guid} gefunden.")

        # Überprüfen, ob das Fenster Assoziationen hat
        if hasattr(ifc_window, "HasAssociations") and ifc_window.HasAssociations:
            print("Assoziationen gefunden:")
            # Alle Assoziationen ausgeben
            for rel in ifc_window.HasAssociations:
                if rel.is_a("IfcRelAssociatesMaterial"):
                    material = rel.RelatingMaterial
                    if material.is_a("IfcMaterialConstituentSet"):
                        print(f"Gefundenes Material Constituent Set mit {len(material.MaterialConstituents)} Materialien:")
                        for constituent in material.MaterialConstituents:
                            print(f"Material Name: {constituent.Material.Name}, Rolle: {constituent.Name}")
        else:
            print("Keine Material-Assoziationen gefunden.")
        break
else:
    print(f"Kein Fenster mit der IfcGuid {target_guid} gefunden.")
```

```python
import ifcopenshell

# IFC-Datei öffnen
ifc_file = ifcopenshell.open("../Hus28_test.ifc")

# Die gesuchte IfcGuid / GlobalId
target_guid = "1wOE2MwBf03f9R8GpskFuS"

# Suche das spezifische Fenster
target_window = None
for ifc_window in ifc_file.by_type("IfcWindow"):
    if ifc_window.GlobalId == target_guid:
        target_window = ifc_window
        break

if target_window:
    print(f"Fenster mit GlobalId {target_guid} gefunden.")
    
    # Durchsuche alle IfcRelAssociatesMaterial nach unserem Fenster
    for rel in target_window.HasAssociations:
        if rel.is_a("IfcRelAssociatesMaterial"):
            material = rel.RelatingMaterial
            if material.is_a("IfcMaterialLayerSetUsage"):
                for layer in material.ForLayerSet.MaterialLayers:
                    print(f"Material Layer: {layer.Material.Name}")
            elif material.is_a("IfcMaterialList"):
                for mat in material.Materials:
                    print(f"Material: {mat.Name}")
            elif material.is_a("IfcMaterialConstituentSet"):
                for constituent in material.MaterialConstituents:
                    print(f"Material: {constituent.Material.Name}, Rolle: {constituent.Name}")
            elif material.is_a("IfcMaterial"):
                print(f"Einzelnes Material: {material.Name}")
else:
    print(f"Kein Fenster mit der IfcGuid {target_guid} gefunden.")
```

```python
import ifcopenshell

# IFC-Datei öffnen
ifc_file = ifcopenshell.open("../Hus28_test.ifc")

# Die gesuchte IfcGuid / GlobalId
target_guid = "1wOE2MwBf03f9R8GpskFuS"

# Suche das spezifische Fenster
target_window = None
for ifc_window in ifc_file.by_type("IfcWindow"):
    if ifc_window.GlobalId == target_guid:
        target_window = ifc_window
        break

if target_window:
    print(f"Fenster mit GlobalId {target_guid} gefunden.")
    
    # Durchsuche alle IfcRelDefinesByType nach unserem Fenster
    for rel_def in ifc_file.by_type("IfcRelDefinesByType"):
        if target_window in rel_def.RelatedObjects:
            window_type = rel_def.RelatingType
            
            # Durchlaufe alle Property Sets des IfcWindowType
            for prop_set in window_type.HasPropertySets:
                if prop_set.is_a("IfcPropertySet"):
                    for prop in prop_set.HasProperties:
                        if prop.is_a("IfcPropertySingleValue"):
                            if prop.Name in ["Material Exterior", "Material Interior", "Material Glass"]:
                                print(f"{prop.Name}: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
            
            # Suche in allen Materialzuweisungen
            for rel in window_type.HasAssociations:
                if rel.is_a("IfcRelAssociatesMaterial"):
                    material = rel.RelatingMaterial
                    if material.is_a("IfcMaterialConstituentSet"):
                        for constituent in material.MaterialConstituents:
                            if constituent.Name in ["Glass", "Material Glass"]:
                                print(f"Material Glass: {constituent.Material.Name}")
                    elif material.is_a("IfcMaterialList"):
                        for mat in material.Materials:
                            if mat.Name in ["Glass", "Material Glass"]:
                                print(f"Material Glass: {mat.Name}")
                    elif material.is_a("IfcMaterialLayerSet"):
                        for layer in material.MaterialLayers:
                            if layer.Material.Name in ["Glass", "Material Glass"]:
                                print(f"Material Glass: {layer.Material.Name}")
else:
    print(f"Kein Fenster mit der IfcGuid {target_guid} gefunden.")
```

```python
import ifcopenshell

# IFC-Datei öffnen
ifc_file = ifcopenshell.open("../Hus28_test.ifc")

# Die gesuchte IfcGuid / GlobalId
target_guid = "1wOE2MwBf03f9R8GpskFuS"

# Suche das spezifische Fenster
target_window = None
for ifc_window in ifc_file.by_type("IfcWindow"):
    if ifc_window.GlobalId == target_guid:
        target_window = ifc_window
        break

if target_window:
    # Durchsuche alle IfcRelDefinesByType nach unserem Fenster
    for rel_def in ifc_file.by_type("IfcRelDefinesByType"):
        if target_window in rel_def.RelatedObjects:
            window_type = rel_def.RelatingType
            
            # Durchlaufe alle Property Sets des IfcWindowType
            for prop_set in window_type.HasPropertySets:
                if prop_set.is_a("IfcPropertySet"):
                    for prop in prop_set.HasProperties:
                        if prop.is_a("IfcPropertySingleValue"):
                            if prop.Name == "Material Exterior":
                                print(f"Material Exterior: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
                            elif prop.Name == "Material Interior":
                                print(f"Material Interior: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
                            elif prop.Name == "Light 1 Operation":
                                print(f"Light 1 Operation: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
                            elif prop.Name == "Light 2 Operation":
                                print(f"Light 2 Operation: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
                            elif prop.Name == "Light 3 Operation":
                                print(f"Light 3 Operation: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
                            elif prop.Name == "Panel Operation":
                                print(f"Panel Operation: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
                            elif prop.Name == "Type":
                                print(f"Type: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
            
            # Zusätzliche Überprüfung auf `Type` in anderen Assoziationen
            if not any(prop.Name == "Type" for prop in prop_set.HasProperties):
                # Durchsuche alle anderen Property Sets des Fensters selbst (nicht nur des Typs)
                for rel_def in target_window.IsDefinedBy:
                    if rel_def.is_a("IfcRelDefinesByProperties"):
                        property_set = rel_def.RelatingPropertyDefinition
                        if property_set.is_a("IfcPropertySet"):
                            for prop in property_set.HasProperties:
                                if prop.is_a("IfcPropertySingleValue") and prop.Name == "Type":
                                    print(f"Type: {getattr(prop.NominalValue, 'wrappedValue', 'N/A')}")
                                    break
            break
else:
    print(f"Kein Fenster mit der IfcGuid {target_guid} gefunden.")
```
